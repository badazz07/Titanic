# Clean the Name column

```python
data['Name']
```

```python
# define a function to extract the title, first name, and last name
def extract_name_components(name):
    # split the name comma to spartae last name from the rest
    last_name, rest_of_name = name.split(',',1)

    # split the rest of the name by spaces to separate title and first name
    name_parts = rest_of_name.strip().split(' ')

    # Extract title
    title = name_parts[0]

    #First name might have multiple parts, so join them except the title
    first_name = ' '.join(name_parts[1:])

    # Clean up any extra characters from first name
    first_name = first_name.replace('(', '').replace(')', '')

    return title, first_name, last_name.strip()
```

```python
def extract_name_components(name):
    last_name, rest_of_name = name.split(',',1)
    name_parts = rest_of_name.strip().split(' ')
    title = name_parts[0]
    first_name = ' '.join(name_parts[1:])
    first_name = first_name.replace('(', '').replace(')', '')

    return title, first_name, last_name.strip()
```

```python
# Apply the function to the Name column
data['Title'], data['First Name'], data['Last Name'] = zip(*data['Name'].map(extract_name_components))
```

# Features selection/engineering

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
x = data.drop('Survived', axis = 1) # x here represents our features (independent variable)
y = data['Survived'] # y here represents our label (dependent variable)
x_train, x_test, y_train,y_test = train_test_split(x,y, test_size=0.20, random_state=30) #test size must not be more than 0.20 (20%) so it can train more
#also we can write it this way. Random state is to pick the data at random or to shuffle 
#x = data[['Pclass','Sex','Age','Number_of_siblings', 'Parch','Fare','Embarked']]
#y = data['Survived ']
```

```python
x_train
```

```python
y_train
```

```python
lr = LogisticRegression() # We make an obj of LR algorithm
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
y_pred
```

```python
pd.set_option('display.max_rows', None)
```

```python
y_train.shape 
```

```python
y_test
```

```python
# Evaluate the model using accuracy score, classification report, and confusion matrix
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
print("Classification Report:")
print(classification_report(y_test, y_pred))
print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred))
```
